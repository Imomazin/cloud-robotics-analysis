```python
# Strategic AI Systems: Data Analysis and Visualization Notebook
# This notebook reproduces the analysis conducted for the paper
# "Strategic AI Systems: Evolution and Implementation Frameworks"

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from collections import Counter
import nltk
from nltk.corpus import stopwords
import networkx as nx
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
import plotly.express as px
import plotly.graph_objects as go

# Set visualization styles
# Fix for newer versions of matplotlib where 'seaborn-whitegrid' is deprecated
try:
    plt.style.use('seaborn-whitegrid')
except:
    # For matplotlib 3.6 and newer
    plt.style.use('seaborn-v0_8-whitegrid')

sns.set_palette("viridis")

# Create sample data since GitHub URLs aren't available yet
# We'll create the data in memory first, then you can upload it to GitHub later
def create_sample_data():
    """
    Create sample data that can be used for the analysis
    """
    # Sample coded segments data
    sample_segments = pd.DataFrame({
        'source_id': ['doc001', 'doc001', 'doc002', 'doc003', 'doc003', 'doc004', 'doc005'] * 30,
        'code': ['Augmented Analytics', 'Data Integration', 'Scenario Modeling', 'Trust Development',
                'Strategic Option Generation', 'Governance Model', 'Autonomous Recommendation'] * 30,
        'segment': [
            "The system functioned essentially as an extremely intelligent dashboard that could visualize our market position.",
            "We struggled with integrating diverse data sources including unstructured text and external market signals.",
            "The AI system generated scenarios based on combinations of variables we hadn't considered before.",
            "Initial trust was limited, but grew as the system demonstrated its capabilities in lower-stakes applications.",
            "What surprised us was the system's ability to propose options that weren't obvious recombinations of existing strategies.",
            "We needed clear governance frameworks defining who had decision authority in different strategic contexts.",
            "The most advanced implementations incorporated organizational history and values into recommendations."
        ] * 30
    })

    # Sample metadata
    sample_metadata = pd.DataFrame({
        'source_id': ['doc001', 'doc002', 'doc003', 'doc004', 'doc005'] * 20,
        'source_type': ['Case Study', 'Technical Document', 'Executive Interview', 'Case Study', 'Technical Document'] * 20,
        'publication_year': [2022, 2023, 2024, 2023, 2024] * 20,
        'organization_sector': ['Financial Services', 'Technology', 'Retail', 'Manufacturing', 'Technology'] * 20,
        'implementation_stage': ['Stage 1', 'Stage 2', 'Stage 3', 'Stage 2', 'Stage 4'] * 20
    })

    # Create additional codes and segments for more diversity in the analysis
    additional_codes = [
        'Data Quality', 'Trust Barrier', 'Technical Complexity', 'Explainability Issue',
        'Knowledge Representation', 'Learning System', 'Professional Role', 'Process Integration',
        'Governance Framework', 'Trust Development', 'Process Redesign', 'Explainable AI',
        'Data Governance', 'Professional Development', 'Integration Challenge'
    ]

    additional_segments = [
        "Data quality was a significant challenge for strategic applications.",
        "Trust barriers were highest in early implementations.",
        "Technical complexity increased with each evolutionary stage.",
        "Explainability was essential for executive adoption.",
        "Knowledge representation frameworks evolved from generic to organization-specific.",
        "Learning systems required careful design to capture strategic outcomes.",
        "Professional roles needed explicit redefinition as AI capabilities advanced.",
        "Process integration was more successful when existing processes were redesigned.",
        "Governance frameworks clarified decision rights across human-AI boundaries.",
        "Trust development followed a predictable pattern across implementations.",
        "Process redesign yielded better results than appending AI to existing processes.",
        "Explainable AI mechanisms were tailored to executive audiences.",
        "Data governance for strategic applications differed from operational systems.",
        "Professional development programs focused on complementary human capabilities.",
        "Integration challenges were most pronounced in the strategic option generation stage."
    ]

    # Add additional data
    additional_df = pd.DataFrame({
        'source_id': ['doc006', 'doc007', 'doc008', 'doc009', 'doc010'] * 3,
        'code': additional_codes,
        'segment': additional_segments
    })

    # Combine datasets
    sample_segments = pd.concat([sample_segments, additional_df], ignore_index=True)

    additional_metadata = pd.DataFrame({
        'source_id': ['doc006', 'doc007', 'doc008', 'doc009', 'doc010'],
        'source_type': ['Case Study', 'Executive Interview', 'Technical Document', 'Case Study', 'Executive Interview'],
        'publication_year': [2024, 2023, 2024, 2022, 2023],
        'organization_sector': ['Healthcare', 'Financial Services', 'Technology', 'Professional Services', 'Retail'],
        'implementation_stage': ['Stage 2', 'Stage 3', 'Stage 4', 'Stage 1', 'Stage 3']
    })

    sample_metadata = pd.concat([sample_metadata, additional_metadata], ignore_index=True)

    return sample_segments, sample_metadata

# Generate sample data
df, metadata = create_sample_data()

print(f"Created sample dataset with {len(df)} coded segments from {df['source_id'].nunique()} documents")
print(f"Created sample metadata for {len(metadata)} documents")

# Display dataset information
df.info()
df.head()

# Analyze distribution of sources by type and year
def visualize_source_distribution():
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

    # Distribution by source type
    source_counts = metadata['source_type'].value_counts()
    sns.barplot(x=source_counts.index, y=source_counts.values, ax=ax1)
    ax1.set_title('Distribution of Sources by Type')
    ax1.set_ylabel('Count')
    ax1.set_xticklabels(ax1.get_xticklabels(), rotation=45, ha='right')

    # Distribution by year
    year_counts = metadata['publication_year'].value_counts().sort_index()
    sns.barplot(x=year_counts.index, y=year_counts.values, ax=ax2)
    ax2.set_title('Distribution of Sources by Publication Year')
    ax2.set_ylabel('Count')
    ax2.set_xlabel('Publication Year')

    plt.tight_layout()
    plt.savefig('source_distribution.png', dpi=300)
    plt.show()

visualize_source_distribution()

# Analyze code frequency and distribution
def analyze_code_distribution():
    # Count code frequency
    code_counts = df['code'].value_counts().reset_index()
    code_counts.columns = ['Code', 'Frequency']

    # Visualize top 20 codes
    plt.figure(figsize=(12, 8))
    sns.barplot(x='Frequency', y='Code', data=code_counts.head(20))
    plt.title('Top 20 Codes by Frequency')
    plt.tight_layout()
    plt.savefig('top_codes.png', dpi=300)
    plt.show()

    # Analyze code distribution across source types
    code_by_source = df.merge(metadata[['source_id', 'source_type']], on='source_id')
    pivot = pd.crosstab(code_by_source['code'], code_by_source['source_type'])

    # Normalize to percentage
    pivot_pct = pivot.div(pivot.sum(axis=1), axis=0) * 100

    # Get top 15 codes
    top_codes = code_counts['Code'].head(15)
    pivot_pct_top = pivot_pct.loc[top_codes]

    # Visualize distribution
    plt.figure(figsize=(12, 8))
    sns.heatmap(pivot_pct_top, annot=True, cmap='viridis', fmt='.1f')
    plt.title('Distribution of Top 15 Codes Across Source Types (%)')
    plt.tight_layout()
    plt.savefig('code_distribution.png', dpi=300)
    plt.show()

    return code_counts

code_counts = analyze_code_distribution()

# Create co-occurrence network of codes
def create_cooccurrence_network():
    # Group by document and get all codes used in each document
    doc_codes = df.groupby('source_id')['code'].apply(list)

    # Create network
    G = nx.Graph()

    # Add nodes for each code
    for code in df['code'].unique():
        G.add_node(code)

    # Add edges for co-occurring codes
    for codes in doc_codes:
        for i, code1 in enumerate(codes):
            for code2 in codes[i+1:]:
                if G.has_edge(code1, code2):
                    G[code1][code2]['weight'] += 1
                else:
                    G.add_edge(code1, code2, weight=1)

    return G

# Visualize the co-occurrence network
def visualize_network(G):
    # Filter to include only edges with weight > 2
    H = G.copy()
    for u, v, d in list(H.edges(data=True)):
        if d['weight'] <= 2:
            H.remove_edge(u, v)

    # Remove isolated nodes
    H.remove_nodes_from(list(nx.isolates(H)))

    # Calculate node positions using force-directed layout
    pos = nx.spring_layout(H, k=0.3, iterations=100, seed=42)

    # Get top 30 nodes by degree
    top_nodes = sorted(H.degree, key=lambda x: x[1], reverse=True)[:30]
    top_node_names = [n for n, d in top_nodes]

    # Subgraph of top nodes
    H_sub = H.subgraph(top_node_names)
    pos_sub = {node: pos[node] for node in H_sub.nodes()}

    # Calculate edge weights and node sizes
    edge_weights = [H_sub[u][v]['weight'] for u, v in H_sub.edges()]
    normalized_weights = [w/max(edge_weights)*5 for w in edge_weights]
    node_sizes = [H_sub.degree(node)*20 for node in H_sub.nodes()]

    # Create visualization
    plt.figure(figsize=(14, 14))
    nx.draw_networkx_nodes(H_sub, pos_sub, node_size=node_sizes, node_color='lightblue', alpha=0.8)
    nx.draw_networkx_edges(H_sub, pos_sub, width=normalized_weights, alpha=0.5, edge_color='gray')
    nx.draw_networkx_labels(H_sub, pos_sub, font_size=10, font_family='sans-serif')

    plt.axis('off')
    plt.title('Co-occurrence Network of Top 30 Codes', fontsize=16)
    plt.tight_layout()
    plt.savefig('code_network.png', dpi=300, bbox_inches='tight')
    plt.show()

    return H_sub

# Generate network
G = create_cooccurrence_network()
H_sub = visualize_network(G)

# Define the stages for analysis
stages = ['Augmented Analytics', 'Scenario Modeling', 'Strategic Option Generation', 'Autonomous Recommendation']

# Analyze evolutionary stages
def analyze_evolutionary_stages():
    stage_data = {}

    # Extract segments related to each stage
    for stage in stages:
        stage_data[stage] = df[df['code'].str.contains(stage, case=False)]
        print(f"{stage}: {len(stage_data[stage])} coded segments")

    # Create a dataframe with stage characteristics
    stage_chars = []

    # Get all unique codes except stage names
    all_codes = set(df['code']) - set(stages)

    # For each stage, calculate code frequencies
    for stage in stages:
        stage_codes = stage_data[stage]['code'].value_counts()
        total = len(stage_data[stage])

        # Calculate percentage of each code within this stage
        for code in all_codes:
            count = stage_codes.get(code, 0)
            pct = (count / total) * 100 if total > 0 else 0
            stage_chars.append({
                'Stage': stage,
                'Code': code,
                'Count': count,
                'Percentage': pct
            })

    # Convert to dataframe
    stage_char_df = pd.DataFrame(stage_chars)

    # Get top characteristics for each stage
    top_chars = stage_char_df.sort_values('Percentage', ascending=False).groupby('Stage').head(10)

    # Visualize top characteristics for each stage
    plt.figure(figsize=(16, 12))
    for i, stage in enumerate(stages):
        plt.subplot(2, 2, i+1)
        stage_top = top_chars[top_chars['Stage'] == stage].sort_values('Percentage', ascending=False)
        sns.barplot(x='Percentage', y='Code', data=stage_top)
        plt.title(f"Top Characteristics: {stage}")
        plt.tight_layout()

    plt.savefig('stage_characteristics.png', dpi=300, bbox_inches='tight')
    plt.show()

    return stage_char_df, stage_data

# Analyze evolutionary stages
stage_char_df, stage_data = analyze_evolutionary_stages()

# Create visualizations for the implementation factors
def analyze_implementation_factors():
    # Define the main implementation factor categories
    factors = [
        'Data Integration', 'Explainability', 'Knowledge Representation',
        'Trust Development', 'Professional Role', 'Governance Model',
        'Process Integration', 'Learning System'
    ]

    # Extract segments related to each factor
    factor_data = {}
    for factor in factors:
        factor_data[factor] = df[df['code'].str.contains(factor, case=False)]
        print(f"{factor}: {len(factor_data[factor])} coded segments")

    # Create a heatmap of factors across stages
    factor_stage_counts = pd.DataFrame(index=factors, columns=stages, data=0)

    for factor in factors:
        for stage in stages:
            # Count segments that mention both factor and stage
            count = len(df[(df['code'].str.contains(factor, case=False)) &
                           (df['segment'].str.contains(stage, case=False))])
            factor_stage_counts.loc[factor, stage] = count

    # Normalize to percentages
    factor_stage_pct = factor_stage_counts.div(factor_stage_counts.sum(axis=1), axis=0) * 100

    # Visualize as heatmap
    plt.figure(figsize=(12, 8))
    sns.heatmap(factor_stage_pct, annot=True, cmap='viridis', fmt='.1f')
    plt.title('Distribution of Implementation Factors Across Evolutionary Stages (%)')
    plt.tight_layout()
    plt.savefig('factor_stage_heatmap.png', dpi=300)
    plt.show()

    return factor_stage_pct

# Analyze implementation factors
factor_stage_pct = analyze_implementation_factors()

# Generate the integrated sociotechnical framework visualization
def create_framework_visualization():
    # Define the components of the framework
    tech_nodes = ['Knowledge Management', 'Data Integration', 'Reasoning Engines',
                  'Interface Design', 'Learning Mechanisms']
    org_nodes = ['Governance Structures', 'Professional Roles', 'Process Integration',
                'Trust Development']
    human_nodes = ['Judgment', 'Creativity', 'Ethics', 'Contextual Understanding']
    outcome_nodes = ['Decision Quality', 'Implementation Effectiveness', 'Learning Adaptation']

    # Create graph
    F = nx.Graph()

    # Add nodes with category attributes
    for node in tech_nodes:
        F.add_node(node, category='Technical')
    for node in org_nodes:
        F.add_node(node, category='Organizational')
    for node in human_nodes:
        F.add_node(node, category='Human')
    for node in outcome_nodes:
        F.add_node(node, category='Outcome')
    for node in stages:
        F.add_node(node, category='Stage')

    # Add central "Strategic Alignment" node
    F.add_node('Strategic Alignment', category='Central')

    # Add edges
    # Connect all technical nodes to Strategic Alignment
    for node in tech_nodes:
        F.add_edge(node, 'Strategic Alignment', weight=2)

    # Connect all organizational nodes to Strategic Alignment
    for node in org_nodes:
        F.add_edge(node, 'Strategic Alignment', weight=2)

    # Connect all human nodes to Strategic Alignment
    for node in human_nodes:
        F.add_edge(node, 'Strategic Alignment', weight=2)

    # Connect all outcome nodes to Strategic Alignment
    for node in outcome_nodes:
        F.add_edge(node, 'Strategic Alignment', weight=2)

    # Connect stages in sequence
    for i in range(len(stages) - 1):
        F.add_edge(stages[i], stages[i+1], weight=3)

    # Connect all stages to Strategic Alignment
    for node in stages:
        F.add_edge(node, 'Strategic Alignment', weight=1.5)

    # Add some cross-category connections
    F.add_edge('Knowledge Management', 'Professional Roles', weight=1)
    F.add_edge('Interface Design', 'Trust Development', weight=1)
    F.add_edge('Reasoning Engines', 'Judgment', weight=1)
    F.add_edge('Learning Mechanisms', 'Learning Adaptation', weight=1)
    F.add_edge('Governance Structures', 'Ethics', weight=1)
    F.add_edge('Process Integration', 'Implementation Effectiveness', weight=1)

    # Set positions for visualization
    pos = {}

    # Position categories in different areas
    # Technical nodes at top
    for i, node in enumerate(tech_nodes):
        pos[node] = np.array([i*2, 4])

    # Organizational nodes at right
    for i, node in enumerate(org_nodes):
        pos[node] = np.array([8, 3-i])

    # Human nodes at left
    for i, node in enumerate(human_nodes):
        pos[node] = np.array([0, 3-i])

    # Outcome nodes at bottom
    for i, node in enumerate(outcome_nodes):
        pos[node] = np.array([2+i*2, 0])

    # Stage nodes in arc at top-right
    stage_positions = [(6, 4), (7, 3.5), (8, 3), (9, 2.5)]
    for i, node in enumerate(stages):
        pos[node] = np.array(stage_positions[i])

    # Strategic Alignment at center
    pos['Strategic Alignment'] = np.array([4, 2])

    # Get node categories for coloring
    node_categories = nx.get_node_attributes(F, 'category')
    category_colors = {
        'Technical': 'skyblue',
        'Organizational': 'lightgreen',
        'Human': 'salmon',
        'Outcome': 'mediumpurple',
        'Stage': 'gold',
        'Central': 'white'
    }
    node_colors = [category_colors[node_categories[node]] for node in F.nodes()]

    # Get edge weights for width
    edge_weights = [F[u][v]['weight'] for u, v in F.edges()]

    # Create visualization
    plt.figure(figsize=(14, 10))
    nx.draw_networkx_nodes(F, pos, node_size=1500, node_color=node_colors, alpha=0.8, edgecolors='black')
    nx.draw_networkx_edges(F, pos, width=edge_weights, alpha=0.5, edge_color='gray')
    nx.draw_networkx_labels(F, pos, font_size=9, font_weight='bold')

    plt.title('Integrated Sociotechnical Framework for Strategic AI Implementation', fontsize=16)
    plt.axis('off')
    plt.tight_layout()
    plt.savefig('framework_visualization.png', dpi=300, bbox_inches='tight')
    plt.show()

    return F

# Create framework visualization
F = create_framework_visualization()

# Create a publication-ready version of Table 1
def create_implementation_factors_table():
    # Define the table data structure
    table_data = {
        'Implementation Factor': [
            'Data Integration Architecture',
            'Explainability Mechanisms',
            'Strategic Knowledge Representation',
            'Trust Development Approach',
            'Professional Role Evolution',
            'Governance Model',
            'Strategic Process Integration',
            'Learning System Design'
        ],
        'Stage 1 Manifestation': [
            'Integration of structured internal data with limited external sources',
            'Basic pattern explanation and visualization',
            'Domain-specific analytical models',
            'Initial exposure in low-stakes contexts',
            'Analytical support for existing roles',
            'Limited integration with existing authorities',
            'Augmentation of environmental analysis phase',
            'Performance feedback on analytical accuracy'
        ],
        'Stage 4 Manifestation': [
            'Comprehensive integration of structured, unstructured, and alternative data',
            'Strategic narrative generation with causal explanations',
            'Organization-specific strategic knowledge models',
            'Calibrated trust with appropriate reliance patterns',
            'Collaborative strategic development with AI as partner',
            'Formal frameworks for human-AI decision authority',
            'Redesigned strategic processes with AI at core',
            'Multi-dimensional learning from strategic outcomes'
        ],
        'Key Success Factors': [
            'Strategic data governance, quality over quantity',
            'Audience-appropriate explanations, confidence communication',
            'Balance between universal and organization-specific concepts',
            'Progressive trust development, explicit calibration',
            'Emphasis on human comparative advantages, explicit boundaries',
            'Clear decision rights, documented rationales',
            'Integration at appropriate points, aligned formats',
            'Balanced stability and adaptation, structured feedback'
        ]
    }

    # Create pandas DataFrame
    implementation_table = pd.DataFrame(table_data)

    # Save to CSV
    implementation_table.to_csv('implementation_factors_table.csv', index=False)

    # Display the table
    from IPython.display import display
    display(implementation_table)

    return implementation_table

# Create the implementation factors table
implementation_table = create_implementation_factors_table()
print("Implementation Factors Table created and saved.")

print("Analysis complete. All visualizations and data files have been created.")

# Create export files for GitHub repository
def save_results():
    # Save summary statistics
    summary = {
        'Total Documents': metadata['source_id'].nunique(),
        'Total Coded Segments': len(df),
        'Unique Codes': df['code'].nunique(),
        'Document Types': metadata['source_type'].value_counts().to_dict(),
        'Publication Years': metadata['publication_year'].value_counts().to_dict()
    }
    summary_df = pd.DataFrame([summary])
    summary_df.to_csv('summary_statistics.csv')

    # Save code counts
    code_counts.to_csv('code_frequency.csv', index=False)

    # Save stage characteristics
    stage_char_df.to_csv('stage_characteristics.csv', index=False)

    # Save implementation factor analysis
    factor_stage_pct.to_csv('implementation_factors_by_stage.csv')

    # Save the sample data for others to use
    df.to_csv('strategic_ai_coded_segments.csv', index=False)
    metadata.to_csv('strategic_ai_documents_metadata.csv', index=False)

    print("All analysis results saved to CSV files.")

# Save results
save_results()

print("Repository setup complete. All files have been created for upload to GitHub.")
```
